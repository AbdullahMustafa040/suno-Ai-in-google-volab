```python
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "0"


from IPython.display import Audio
import nltk  # we'll use this to split into sentences
import numpy as np

from bark.generation import (
    generate_text_semantic,
    preload_models,
)
from bark.api import semantic_to_waveform
from bark import generate_audio, SAMPLE_RATE
```

```python
script = """
Happy Birthday, Tinku Jee, it's your special day,
With your watch by your side, time won't slip away,
You're a unique soul, in every single way,
Your love for reading brightens up the gray.

So let's celebrate you, in a grand style,
With a book in your hand, and a joyful smile,
No wasting time today, let's go that extra mile,
For Tinku Jee, our celebrations compile.


""".replace("\n", " ").strip()
```

```python
script
```

```python
import nltk
nltk.download('punkt')
```

```python
sentences = nltk.sent_tokenize(script)
```

```python
SPEAKER = "v2/en_speaker_2"
silence = np.zeros(int(0.25 * SAMPLE_RATE))  # quarter second of silence

pieces = []
for sentence in sentences:
    audio_array = generate_audio(sentence, history_prompt=SPEAKER)
    pieces += [audio_array, silence.copy()]
```

```python
Audio(np.concatenate(pieces), rate=SAMPLE_RATE)
```
